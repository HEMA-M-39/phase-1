```python
import numpy as np
import matplotlib.pyplot as plt

#  Simulate a binomial distribution
data = np.random.binomial(n=10, p=0.5, size=1000)

#  Calculate and display the mean
mean_value = np.mean(data)
print(f"Mean of simulated data: {mean_value:.2f}")

# Plot the histogram
plt.hist(data, bins=range(12), density=True, alpha=0.7, color='skyblue', edgecolor='black')

# Add labels and title
plt.title('Binomial Distribution (n=10, p=0.5, 1000 experiments)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability Density')
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)
plt.show()

#  Interpretation
if np.isclose(mean_value, 5, atol=0.2):
    print("Matches binomial expectation (n × p = 5).")
else:
    print("Slight deviation due to randomness.")
```
