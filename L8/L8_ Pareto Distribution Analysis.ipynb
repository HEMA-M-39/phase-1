```python
import numpy as np
import matplotlib.pyplot as plt


data = np.random.pareto(a=2, size=1000)

#  Plot the histogram (log scale for better visibility)
plt.hist(data, bins=50, density=True, log=True, color='skyblue', edgecolor='black')
plt.title('Pareto Distribution (Shape = 2, Size = 1000)')
plt.xlabel('Value')
plt.ylabel('Probability Density (Log Scale)')
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)
plt.show()


# Sort data in descending order
sorted_data = np.sort(data)[::-1]

# Calculate total sum of all values
total_sum = np.sum(sorted_data)

# Find the contribution of the top 20% of values
top_20_percent_sum = np.sum(sorted_data[:200])

# Percentage contribution of top 20% of values
contribution = (top_20_percent_sum / total_sum) * 100

print(f"Top 20% of values contribute: {contribution:.2f}% of the total sum")

#  Interpretation
if 75 <= contribution <= 85:
    print("Top 20% of values ≈ 80% of total sum, confirming the 80/20 rule.")
else:
    print("Deviation due to randomness but demonstrates a skewed distribution.")
```
