```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
df=pd.read_csv("online_retail.csv")
high_order = df[df['Quantity'] > 50]
high_order_count = high_order.shape[0]
total_orders = df.shape[0]
p = high_order_count / total_orders
n = total_orders 
k = high_order_count
binom_prob = stats.binom.pmf(k, n, p)
print(f"P(high order) = {binom_prob:.2f}")
x = range(0, n+1)
y = stats.binom.pmf(x, n, p)
plt.vlines(x, 0, y, colors='b', lw=5)
plt.title('Binomial Distribution of High Orders')
plt.xlabel('Number of High Orders')
plt.ylabel('Probability')
plt.show()
```
